```python
def File_to_label(file_name, number_of_notes_per_example, print_length = False):
    file = open(f"{file_name}", "r")
    tex_code = ''
    remove = []
    notes_per = number_of_notes_per_example 
    k = 0
    tot = 0
    total = {}
    s = file.read()
    file.close()
    
    s_tot = r.findall("\((.*?)\)", s)

    for i in range(len(r.findall("\((.*?)\)", s))):
        s_tot[i] = r.sub(r"[^a-zA-Z]+", "", s_tot[i])
        if s_tot[i] == '':
            remove.append(i)
    for i in remove:
        s_tot.pop(i - k)
        k += 1
    tot = 0
    for i in range(len(notes_per)):
        tot += (notes_per[i])
    
    if tot == len(s_tot):
        for j in range(notes_per[0]):
            tex_code += f'({s_tot[j]})'
        total[1] = tex_code
        tex_code = ''
        l2 = notes_per[0]+notes_per[1]
        l1 = notes_per[0]
        for n in range(len(notes_per) - 1):

            if n > 0:
                l2 += notes_per[n+1]
                l1 += notes_per[n]  
            total[n+2] = ''
            for j in range(l1,l2):
                total[n+2] += f'({s_tot[j]})'
                
                
        return total
        
    else:      
        print(f'invalid splitting, total should add to: {len(s_tot)}. You\'re at: {tot}')            
        return
```

```python
def File_to_label_test(file_name, number_of_notes_per_example):
    file = open(f"{file_name}", "r")
    remove, s_spef, dict_assign = [], [], []
    notes_per = number_of_notes_per_example 
    k = 0
    tot = 0
    dict_total = {}
    file_title = r.sub("\\..*", "", file_name)
    s = file.read()
    file.close()
    
    s_tot = r.findall("\((.*?)\)", s)

    for i in range(len(r.findall("\((.*?)\)", s))):
        s_tot[i] = r.sub(r"[^a-zA-Z]+", "", s_tot[i])
        if s_tot[i] == '':
            remove.append(i)
    for i in remove:
        s_tot.pop(i - k)
        k += 1
    s_tot.pop(0)
    tot = 0
    for i in range(len(notes_per)):
        tot += (notes_per[i])
    
    if tot == len(s_tot):
        for j in range(notes_per[0]):
            s_spef.append(s_tot[j])
            
        dict_total[0] = s_spef.copy()
        s_spef.clear()
        
        l2 = notes_per[0]+notes_per[1]
        l1 = notes_per[0]
        
        for n in range(len(notes_per) - 1):
            if n > 0:
                l2 += notes_per[n+1]
                l1 += notes_per[n]  
            
            for j in range(l1,l2):
                s_spef.append(s_tot[j])
            dict_total[n+1] = s_spef.copy()
            s_spef.clear()
                
                
        return dict_total
        
    else:
        print(f'For {file_name} invalid splitting, total should add to: {len(s_tot)}. You\'re at: {tot}')            
        return

# File_to_label(file name(gabc for an example), a list with how many notes are in each jpeg)
# Output: Dictionary[the example[name of file and index(starts at 1, ends at n)][Gabc code][target, our Y[i,:]]
```

```python
def File_to_label_for_sep_double_notes(file_name, number_of_notes_per_example):
    file = open(f"{file_name}", "r")
    remove, s_spef, dict_assign, s_spef_1, s_spef_2 = [], [], [], [], []
    notes_per = number_of_notes_per_example 
    k = 0
    tot = 0
    dict_total = {}
    file_title = r.sub("\\..*", "", file_name)
    s = file.read()
    file.close()
    
    s_tot = r.findall("\((.*?)\)", s)

    for i in range(len(r.findall("\((.*?)\)", s))):
        s_tot[i] = r.sub(r"[^a-zA-Z]+", "", s_tot[i])
        if s_tot[i] == '':
            remove.append(i)
    for i in remove:
        s_tot.pop(i - k)
        k += 1
    s_tot.pop(0)
    tot = 0
    for i in range(len(notes_per)):
        tot += (notes_per[i])
    
    if tot == len(s_tot):
        for j in range(notes_per[0]):
            if len(s_tot[j]) == 2:
                s_spef_2.append(s_tot[j])
            else: 
                s_spef_1.append(s_tot[j])
            
        dict_total[0] = [s_spef_1.copy(),s_spef_2.copy()]
        s_spef_1.clear()
        s_spef_2.clear()
        
        l2 = notes_per[0]+notes_per[1]
        l1 = notes_per[0]
        
        for n in range(len(notes_per) - 1):
            if n > 0:
                l2 += notes_per[n+1]
                l1 += notes_per[n]  
            
            for j in range(l1,l2):
                if len(s_tot[j]) == 2:
                    s_spef_2.append(s_tot[j])
                else: 
                    s_spef_1.append(s_tot[j])
            dict_total[n+1] = [s_spef_1.copy(),s_spef_2.copy()]
            s_spef_1.clear()
            s_spef_2.clear()
                
                
        return dict_total
        
    else:
        print(f'For {file_name} invalid splitting, total should add to: {len(s_tot)}. You\'re at: {tot}')            
        return
    
# File_to_label(file name(gabc for an example), a list with how many notes are in each jpeg)
# Output: Dictionary[the example[name of file and index(starts at 1, ends at n)][Y[i,:]]
# We want to compare: Algoritm[file][i'th note] = Dict[Gabc of file][i'th note label]
# To be fair we can actually ignore the Gabc code as theres no way to compile it in notebook, adjust code to reflect this
```
